```python
class Node:

    # initializing data and next

    def __init__(self, data: any):

        if data is not None:
            self.data: any = data 
            self.next = None
        else:
            raise TypeError ("Provide Valid Input!")

    # getters
    def get_data(self):
        if self.data:
            return self.data
        else:
            print("The list is empty!")
    
    def get_next(self):
        if self.next:
            return self.next
        else:
            print("We are at the tail!")

    # Setters
    def set_data(self, val: any):
        if val is not None:
            self.data = val
            return self.data
        else:
            raise TypeError ("Provide Valid Input.")
    
    def set_next(self, val: any):
        self.next = val
        return self.next


class LinkedList (Node):

    # Initializing state
    def __init__(self, head: any = None):
        self.head = head
        self.counter = 0

    def inisert(self, val):
        if val:
            new_node = Node(val)

            # changing next
            new_node.set_next(self.head)
            
            # changing head
            self.head = new_node

            self.counter += 1

    def __repr__(self) -> str:
        collected_elements= [

        ]
        
        current = self.head

        while current is not None:

            if current == self.head:
                collected_elements.append( '[Head: %s]' %current.get_data() )

            elif current.get_next() == None:
                collected_elements.append( '[Tail: %s]' %current.get_data() )
                
            else:
                collected_elements.append( '[%s]' %current.get_data() )

            current = current.get_next()
        
        return ' -> '.join(collected_elements)
    
element = LinkedList()
element.inisert(1)
element.inisert(2)
element.inisert(3)
element.inisert(4)
element.inisert(5)
element.inisert(6)


element

    
```

```python
'''
Create a recursive function and compare the mid of the search space with the target. 
And based on the result either return the index where the target is found or call the 
recursive function for the next search space.
'''
def recursive_binarry_search(elements: list, target: any, counter: int = 0):
    if len(elements) == 0:
        return -1
    else:
        midpoint = (len(elements) // 2)

        if elements[midpoint] == target:
            counter+=midpoint

            return counter, True
        
        elif elements[midpoint] < target:
            return recursive_binarry_search(elements[midpoint + 1: ], target, counter = counter + midpoint)
        else:
            return recursive_binarry_search(elements[ : midpoint], target, counter)

values= list(range(1, 100))
target = 340

caller = recursive_binarry_search(values, target)

print(caller)

""" 
Question: (Linear Binary Seach)
Implement a binary search algorithm in Python to find the position of a specific element 35 in a sorted 
array of integers 
[10, 20, 30, 35, 40, 50, 60, 70, 80]. 
If the element is found, return its index; otherwise, return -1.
"""

def linear_binary_search(elements: list, target = 35):
    first_possition = 0
    last_possition = (len(elements) - 1) 

    while first_possition <= last_possition:
        

        midpoint = ((first_possition + last_possition) // 2)

        if elements[midpoint] == target:
            return midpoint
            
        elif elements[midpoint] < target:
            first_possition = midpoint + 1

        else:
            last_possition = midpoint - 1
    
    return -1

l_values= [10, 20, 30, 35, 40, 50, 60, 70, 80]


l_caller = linear_binary_search(l_values)

print(l_caller)
```

```python
'''
Create a binarry function and compare the mid of the search space with the target. 
And based on the result either return the index where the target is found or call the 
recursive function for the next search space.
'''
def binary_search_alg(values : list, target : int = 60):

    first_index = 0
    last_index = (len(values) - 1)

    while first_index <= last_index:
        midpoint = (first_index + last_index) // 2

        if values [midpoint] == target:
            return midpoint
        elif values[midpoint] < target:
            first_index = midpoint + 1
        else:
            last_index = midpoint -1

    return -1

_values= [10, 20, 30, 35, 40, 50, 60, 70, 80]
_caller = binary_search_alg(_values)

print(_caller)

def rec_binary_search(values: list, target: any, counter: int = 0):
    
    if len(values) == 0:
        return -1
    
    else:
        midpoint = (len(values) // 2)

        if values[midpoint] == target:
            counter += midpoint
            return True, counter
        
        elif values[midpoint] < target:
            return rec_binary_search(values[midpoint + 1: ], target, counter = counter + midpoint+1)
        
        else:
            return rec_binary_search(values[ :midpoint], target, counter)

N_values= list(range(1, 100+1))
N_target = 40

N_caller = rec_binary_search(N_values, N_target)

print(N_caller)
```
