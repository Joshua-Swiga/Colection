```python
# Linked Lists- A linear data structure where each is contained in a seperate object called a node

"""
Each Data structure solves a perticular problem.
The first object of a node is called a 'Head', while the last is called a 'Tail'

[Head-> 1 -> |memory_location_of_next|, 2-> |memory_location_of_next|, 3 ->|memory_location_of_next|, 4 Tail] 

Here, each element points to the next:
2 to 3, 3 to 4.
"""
# Constracting a node class. 
class Node:
    # Class attributes
    data = None 
    
    next_node = None

    def __init__(self, data):
        self.data = data

    def __repr__(self) -> str:
        return f'{self.data}'




class Linked_List:

    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None
    
    def size(self):
        
        # returns the number of nodes in the list

        current = self.head
        count = 0

        while current:
            count += 1

            current = current.next_node

        return count
    
    def add(self, data):
        # Adds a new node containing data at the head of the list; taking O(n)
        
        new_node = Node (data)
        new_node.next_node = self.head
        self.head = new_node
```

In the `Node` class of a linked list implementation, the `next_node` attribute is initialized to `None` because initially, when a node is created, it does not point to any other node. By setting `next_node` to `None`, we indicate that this node is the last node in the list or the only node if it's a singleton list.

The `next_node` attribute holds a reference (or pointer) to the next node in the linked list. In a singly linked list, each node contains a reference to the next node in the sequence. By following these references, you can traverse the entire linked list, starting from the head node.

Here's a brief explanation of how it works:

- When you create a new node, its `next_node` attribute is initialized to `None` because it doesn't point to any other node yet.
- When you add this node to the linked list, you update the `next_node` attribute of the previous node to point to this new node. This effectively links the new node into the list.
- If you repeat this process for multiple nodes, each node's `next_node` attribute will point to the next node in the sequence, forming the linked list structure.

So, in summary, `next_node` stores the reference to the next node in the linked list, allowing you to traverse the list from one node to the next. Initially, it's `None` for the last node in the list, and it's updated to point to the next node as new nodes are added to the list.


The methods defined in the `MyNode` class, such as `get_value`, `get_next`, `set_value`, and `set_next`, are intended to be used internally within the class itself or within other classes that interact with nodes (such as a `LinkedList` class).

When you create a linked list using the `LinkedList` class, you'll use methods like `insert`, `append`, `delete`, etc., to manipulate the list. These methods will internally create and manipulate nodes by using the methods provided by the `MyNode` class.

For example, when you insert a new element into the linked list, the `insert` method of the `LinkedList` class will create a new node object using the `MyNode` class constructor (`__init__`). Then, it will use the setter methods (`set_value` and `set_next`) of the `MyNode` class to set the value and the reference to the next node of the newly created node.

Similarly, when you want to retrieve information from a node (such as its value or the reference to the next node), you'll use the getter methods (`get_value` and `get_next`) provided by the `MyNode` class, which will be called internally within the `LinkedList` class or other classes that work with nodes.

By encapsulating the logic for working with nodes within the `MyNode` class and providing getter and setter methods, you ensure that the internal details of how nodes are implemented are hidden from the external users of the class, promoting encapsulation and abstraction in your code.


```python
class Node:
    def __init__(self, value):
        #Initializing the state of this node: with value and next memory reference
        self.value = value # The head- The first value of the linked list
        self.next = None # This is going to hold the next memory reference

    #getter -- to read the state of the node
    def get_data(self):
        return self.value

    #Allows us to read the next value
    def get_next(self):
        return self.next
    

    #setter -- to change the state of a the node
    def set_data(self, new_value):
        self.value = new_value
        
    """ 
    since the linked list is initially empty, there is no next node to point to. 
    Therefore, when creating the first node, we initialize the next_node attribute to None 
    to indicate that there is no next node.As the linked list grows and additional nodes are added, 
    the next_node attribute of each node is updated to point to the next node in the sequence.
    """

    # Will allow us to set the memory location of the next element
    def set_next (self, next):
        self.next = next
# Next we proceed to create the Linked list

"""
1. create a constractor to provide an initial state
    -we start with two attributes:
    1. head (the first value in the linked list)
    2. count (the number of nodes in the linked list) -- will allow us to access the data as an attribute
2. create funtion to add items to the list

3. create a function that will allow us to itterate over each node O(n)
    -If the element we want us pressent, returns the node
    -else returns None
"""
    
class Linked_list:
    def __init__(self, head = None):
        self.head = head
        self.count = 0 # This will essentially tell us the number of nodes the list has
    

    def insert(self, data):
        """
        It is used to insert a new node containing the provided data at the 
        beginning of the linked list. 
        """
        new_node = Node(data)
        """
        A new node is created. This creates a new node object. 
        self.value = value # The head, is therefore set to 'data'
        self.next = None (reference to the next is set to None)
        """

        new_node.set_next(self.head)
        """ 
        This line of code sets the next reference of the newly created node (new_node) to point to the current head 
        of the linked list (self.head). 
        In other words, we are saying that the next node after the newly 
        created node will be the current head node of the linked list. 
        This establishes the connection between the new node and the rest of the linked list.
        """
        self.head = new_node
        """ 
        This line of code updates the head pointer of the linked list to 
        point to the newly created node (new_node). 
        By doing this, we effectively make the newly created node the new head of the 
        linked list. Now, whenever we want to access the first node in the linked list, 
        we can do so by following the head pointer, 
        which now points to the newly inserted node.
        """
        self.count += 1

    def find (self, value):
        item = self.head

        while item != None:
            if item.get_data() == value:
                return item
            else:
                item = item.get_next()

        return None


    """ 
    Now that we can find elements, how can we delete them?
    """

data_no_1 = "First"
data_no_2 = "Second"
data_no_3 = "Third"
data_no_4 = "Fourth"
data_no_5 = "Fifth"
data_no_6 = "Sixth"


testing = Linked_list()
testing.insert(data_no_1)
testing.insert(data_no_2)
testing.insert(data_no_3)
testing.insert(data_no_4)
testing.insert(data_no_5)
testing.insert(data_no_6)

testing.count


```
