```python
# There are two ways of writing a binary search algorithim. 
#The second is called recursive binarry search


#Unlike the initial function which returned the index, this will return a boolean output

def recursive_binary_search(list_element: list, target: any):

    if len(list_element) == 0: #Checking if the array is empty
        return False
    else:
        midpoint = (len(list_element)) // 2 #Accessing the midpoint of the list


        # First Terminating condition
        if list_element[midpoint] == target:
            return True
        
        else:
            if list_element[midpoint] < target:
                return recursive_binary_search(list_element[midpoint + 1 : ], target)
            else:
                return recursive_binary_search(list_element[ : midpoint], target)

numbers=range(1, 100 + 1)
target=100

answer=recursive_binary_search(numbers, target)

print(answer)



def second_recursive_binary_search(arry: list, target: int):
    if len(arry) == 0:
        return False
    
    else:
        midpoint = (len(arry)) // 2


        # If this code evaluatess to true, the else statement will not run
        if arry[midpoint] == target:
            return True

        else:
            if arry[midpoint] < target: # Discards everything before the midpoint, including the midpoint
                return second_recursive_binary_search(arry[midpoint + 1 : ], target)
            
            
            else: # Discards everything after the midpoint, including the midpoint
                return second_recursive_binary_search(arry[ : midpoint - 1], target)
            
second_number=range(1, 100 + 1)
second_target=1001

second_answer=second_recursive_binary_search(second_number, second_target)
print(second_answer)
    
```

To apply binary seach:
    1. The elements within the array must be sorted
    2. Acess to any element in the array takes constant time 

The steps involved are as follows:
    1. You need to find the middle index (mid=first_index+last_index) // 2
    2. Compare the middle index to the target element-If the taget maches, terminate the process. 
    3. else if not maching, pick a half for the the next seach space
        -if the target is smaller than the middle element, use the left side
        -if target is larger than the middle element, use the right side. 

Binary Search can be implimented in two ways:
    1. iterative binary search:
        This is where you adjast the fist and last index possition based on whether they are greater or lesser than the target

    2. recursive binary search:
        This is where the function is called recursively in which it takes a new sliced list. It returns a Boolean output indicating whether or
        not the element is pressent in the array

    

```python
'''
Create a recursive function and compare the mid of the search space with the target. 
And based on the result either return the index where the target is found or call the 
recursive function for the next search space.
'''
# A recursive function that is meant to locate an element within an array by comparing it with the middle index

def recursive_binary_func(arry: list, target: int):
    if len(arry) == 0:
        return False
    
    else:
        midpoint = len(arry) // 2

        if arry[midpoint] == target:
            return True
        
        else:
            if arry[midpoint] < target:
                return recursive_binary_func(arry[midpoint + 1:], target)
            else:
                return recursive_binary_func(arry[: midpoint], target)

number=range(1, 100 + 1)
target=18

answer=recursive_binary_func(number, target)

print(answer)



def rec_function(list: list, target: any):

    if len(list) == 0:
        return None
    
    else:
        midpoint= (len(list)) // 2

        if list[midpoint]==target:
            return True
        else:
            if list[midpoint] < target:
                return rec_function(list[midpoint + 1: ], target)
            else:
                return rec_function(list[ :midpoint], target)
list=range(1, 100+1)
target7=78

result3=rec_function(list, target7)

print(result3)
```
