```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
square_feet = [12.0, 15.0, 18, 13, 11.0]
prices =[2000.0, 2500.0, 3000.0, 2200.0, 1800.0]


consolidated_data = np.array([
    square_feet, prices
])
print(
    consolidated_data[0],'\n',
    consolidated_data[1]
    )
```

```python
plt.scatter(consolidated_data[1], consolidated_data[0])
plt.xlabel("House pricing")
plt.ylabel("Square Feet")
plt.title("Number of Bathrooms Against Pricing")
```

```python
slope, intercept, r, p, std_err = linregress(consolidated_data[1], consolidated_data[0])

print(f"Slope: {slope}. \nIntercept: {intercept}. \nCorrelation Coefficient: {r}. \nConclusion of null hypothesis: {p}. \nStandard Error of slope and intercept: {std_err}")

def equation_of_stright_line(x):
    return slope * x + intercept

model = list(map(equation_of_stright_line, consolidated_data[0])) # This creates all the corresponding values y of x based on: y = mx + c
```

```python
plt.scatter(consolidated_data[1], consolidated_data[0])
plt.plot(consolidated_data[1], model)
plt.show()
```
