### Summary

#### Polynomial Regression

- **Purpose**: Polynomial regression is used to model the relationship between a dependent variable \( y \) and an independent variable \( x \) by fitting a polynomial equation to the observed data points.
- **Polynomial Equation**: A polynomial equation of degree \( n \) is represented as:
  \[
  y = m_n x^n + m_{n-1} x^{n-1} + \ldots + m_2 x^2 + m_1 x + c
  \]
  where:
  - \( m_n, m_{n-1}, \ldots, m_1 \) are the coefficients (slopes) for each term.
  - \( c \) is the intercept.

#### Key Concepts

1. **Degree of Polynomial**:
   - The degree of the polynomial is the highest power of \( x \) in the equation.
   - It determines the maximum number of turns or bends the regression line can have.
   - Example: A degree 2 polynomial is \( y = m_2 x^2 + m_1 x + c \).

2. **Design Matrix**:
   - For polynomial regression, the design matrix includes \( x \) raised to the power corresponding to the polynomial degree.
   - Example for degree 2:
     \[
     X = \begin{bmatrix}
     x^2 & x & 1 \\
     \end{bmatrix}
     \]
   - For a dataset with \( x \) values: \( [1, 2, 3] \), the design matrix is:
     \[
     X = \begin{bmatrix}
     1^2 & 1 & 1 \\
     2^2 & 2 & 1 \\
     3^2 & 3 & 1
     \end{bmatrix}
     = \begin{bmatrix}
     1 & 1 & 1 \\
     4 & 2 & 1 \\
     9 & 3 & 1
     \end{bmatrix}
     \]

3. **Fitting the Model**:
   - Use the method of least squares to fit the polynomial to the data, which involves solving the normal equations:
     \[
     \mathbf{M} = (X^T X)^{-1} X^T y
     \]
   - Here, \( \mathbf{M} \) represents the vector of coefficients: \( [m_n, m_{n-1}, \ldots, m_1, c]^T \).

4. **Prediction**:
   - To predict the value of \( y \) for a given \( x \), use the fitted polynomial equation with the obtained coefficients.
   - Example: For \( x = 6 \) and a quadratic polynomial \( y = m_2 x^2 + m_1 x + c \):
     \[
     y_{\text{pred}} = m_2 \cdot 6^2 + m_1 \cdot 6 + c
     \]

### Terminology

- **Independent Variable** (\( x \)): The variable you are using to predict the dependent variable.
- **Dependent Variable** (\( y \)): The variable you are trying to predict.
- **Coefficient** (\( m \)): The slope of each term in the polynomial equation.
- **Intercept** (\( c \)): The constant term in the polynomial equation.
- **Degree**: The highest power of \( x \) in the polynomial equation, indicating the maximum number of turns the curve can have.
- **Design Matrix**: A matrix that includes the independent variable \( x \) raised to the power corresponding to the polynomial degree.
- **Normal Equations**: The equations used to calculate the coefficients of the polynomial using the method of least squares.
- **Least Squares Method**: A method to minimize the sum of the squared differences between the observed and predicted values.
