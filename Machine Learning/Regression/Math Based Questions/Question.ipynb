### Linear Regression Question

**Dataset**: You are given the following dataset for the number of hours studied (independent variable \(X\)) and the scores obtained in an exam (dependent variable \(Y\)):

| Hours Studied (X) | Exam Score (Y) |
|-------------------|----------------|
| 1                 | 50             |
| 2                 | 55             |
| 3                 | 60             |
| 4                 | 70             |
| 5                 | 75             |

**Tasks**:
1. Calculate the slope (\(b_1\)) and intercept (\(b_0\)) of the linear regression line \(Y = b_0 + b_1X\).
2. Use the regression equation to predict the score for a student who studies 6 hours.
3. Calculate the \( R^2 \) value for the regression model.

### Multiple Regression Question

**Dataset**: You are given the following dataset for the number of hours studied (independent variable \(X_1\)), the number of hours slept (independent variable \(X_2\)), and the scores obtained in an exam (dependent variable \(Y\)):

| Hours Studied (X1) | Hours Slept (X2) | Exam Score (Y) |
|--------------------|------------------|----------------|
| 1                  | 6                | 52             |
| 2                  | 7                | 58             |
| 3                  | 8                | 64             |
| 4                  | 5                | 68             |
| 5                  | 9                | 75             |

**Tasks**:
1. Calculate the coefficients (\(b_0\), \(b_1\), \(b_2\)) of the multiple regression equation \(Y = b_0 + b_1X_1 + b_2X_2\).
2. Use the regression equation to predict the score for a student who studies 6 hours and sleeps 7 hours.
3. Calculate the \( R^2 \) value for the multiple regression model.

### Steps for Linear Regression

1. **Calculate the slope (\(b_1\))**:
   \[
   b_1 = \frac{n(\sum XY) - (\sum X)(\sum Y)}{n(\sum X^2) - (\sum X)^2}
   \]
2. **Calculate the intercept (\(b_0\))**:
   \[
   b_0 = \bar{Y} - b_1\bar{X}
   \]
3. **Regression equation**:
   \[
   Y = b_0 + b_1X
   \]
4. **Predict the score for 6 hours**.
5. **Calculate \( R^2 \)**:
   \[
   R^2 = 1 - \frac{SS_{\text{res}}}{SS_{\text{tot}}}
   \]

### Steps for Multiple Regression

1. **Calculate the coefficients** using the normal equation approach or matrices:
   \[
   \mathbf{b} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}
   \]
   Here, \(\mathbf{X}\) is the matrix of independent variables, \(\mathbf{Y}\) is the vector of the dependent variable, and \(\mathbf{b}\) is the vector of coefficients.

2. **Predict the score** for the given values of \(X_1\) and \(X_2\).
3. **Calculate \( R^2 \)** for multiple regression.
