```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22] 
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
coefficiencts = np.polyfit(x, y, 3)
y_hut = np.poly1d(coefficiencts) # Creates polynomial object

x_for_plotting = np.linspace(min(x), max(x), len(x))

print(len(x_for_plotting) == len(x))

print(len(x))
plt.scatter(x, y)
plt.plot(x_for_plotting, y_hut(x_for_plotting))
plt.show()
```
