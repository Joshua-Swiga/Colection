```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
independent_values=['Square_Footage', 'Year_Built', 'Lot_Size', 'Garage_Size']
```

```python
coefficients = np.polyfit(df[independent_values[0]], response_variable, 10)
model = np.poly1d(coefficients)
x_line = np.linspace(min(df[independent_values[0]]), max(df[independent_values[0]]), len(df[independent_values[0]]))

# Basic Scatter Plot
plt.scatter(df[independent_values[0]], response_variable, marker="*")
plt.title(independent_values[0])
plt.ylabel('Price')
plt.xlabel(independent_values[0])
plt.grid(True)

# Regression Curve
plt.plot(x_line, model(x_line))
plt.show()

#print(x_line)
```

SSE: Useful for understanding the total error in the dataset. It is particularly useful when comparing the fit of different models to the same data.
MSE: Useful for comparing the performance of models on different datasets or when the number of observations differs. It helps in understanding the average squared error.
RMSE: Provides a measure of error in the same units as the response variable, making it easier to interpret and communicate.

```python
# Model Tests
# Variance for x
independent_v = df[independent_values[0]]
mean_v = np.mean(df[independent_values[0]])

residual_v = []
for val in independent_v:
    i = val - mean_v
    residual_v.append(i)

squared_v = [x ** 2 for x in residual_v]
added_v = np.sum(squared_v)

var = added_v/len(independent_v)
print(var)
```
