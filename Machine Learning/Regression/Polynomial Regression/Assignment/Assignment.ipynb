```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'data.csv'

if file is not None:
    df = pd.read_csv(file)
    print("File opened")

else:
    raise FileExistsError("The file does not exist!")
```

```python
column = df.columns

list_of_columns = [ ]
for value in column:
    if value == 'Price' or value == 'Neighborhood':
        continue
    else:
        list_of_columns.append(value)

response_variable = df['Price']

print(list_of_columns)
```

```python
# Categorical Conversion
neighborhood = df['Neighborhood']

for location in neighborhood:
    pass
```

```python
index_possition = 0

for index in list_of_columns:
    independent_variable = list_of_columns[index_possition]
    
    plt.scatter(df[independent_variable], response_variable, marker='*')
    plt.title(f'{list_of_columns[index_possition]} Vs Price')
    plt.xlabel(f'{list_of_columns[index_possition]}')
    plt.ylabel("Price")
    plt.grid(True)
    plt.show()

    index_possition += 1
    
```

```python
independent_values=['Square_Footage', 'Year_Built', 'Lot_Size', 'Garage_Size']
```

```python
coefficients = np.polyfit(df[independent_values[0]], response_variable, 3)
model = np.poly1d(coefficients)
x_line = np.linspace(min(df[independent_values[0]]), max(df[independent_values[0]]), len(df[independent_values[0]]))

# Basic Scatter Plot
plt.scatter(df[independent_values[0]], response_variable, marker="*")
plt.title(independent_values[0])
plt.ylabel('Price')
plt.xlabel(independent_values[0])
plt.grid(True)

# Regression Curve
plt.plot(x_line, model(x_line))
plt.show()

#print(x_line)
```

```python
# Model Tests
```

```python
# SSE
y_hut = model(df[independent_values[0]])
residuals = response_variable - y_hut


squared_residuals = [x ** 2 for x in residuals]
sse = np.sum(squared_residuals)
print(sse)
```

```python
# MSE
mean_squared_error = (np.sqrt(sse/len(residuals)))
```
