```python
# Correlation Coefficient
def correlation_coefficient(x, y):
    # Running Validations
    print("Running Validations...")
    
    if len(x) != len(y):
        print("The length of the lists provided are not proportional!")
        return None

    if not isinstance(x, (list, np.ndarray)) or not isinstance(y, (list, np.ndarray)):
        raise TypeError('Please provide lists or numpy arrays of numerical data points')

    n = len(x)
    print(f"Number of data points (n): {n}")
    
    print("Validations Passed...\nProceeding to arithmetic operations.")

    # Calculate the correlation coefficient (r)
    numerator_value = np.cov(x, y, ddof=0)[0, 1] * n
    print(f"The Numerator Value is: {numerator_value}")

    left_denominator_value = np.sqrt(np.var(x, ddof=0) * n)
    right_denominator_value = np.sqrt(np.var(y, ddof=0) * n)
    denominator_value = left_denominator_value * right_denominator_value
    print(f"The Denominator Value is: {denominator_value}")
    
    if denominator_value == 0:
        print("Denominator is zero. Cannot divide by zero.")
        return None

    answer = numerator_value / denominator_value
    print(f"The Correlation Coefficient (r) is: {answer}")
    return answer

def correlation_coefficient_checker(r_value, independent_variable, dependent_variable=None):
    if r_value == 0:
        print(f"There is no linear relationship between: {independent_variable} and {dependent_variable}")
        return

    elif np.abs(r_value) > 0.1 and np.abs(r_value) <= 0.9:
        print(f"{independent_variable} and {dependent_variable} show evidence of linearity")
        return
    
    elif np.abs(r_value) == 1:
        print(f"{independent_variable} and {dependent_variable} have a very strong linear relationship.")
        return
    elif np.abs(r_value) > 0.9:
        print(f"{independent_variable} and {dependent_variable} have a strong linear relationship.")
        return

    else:
        if r_value < 0:
            print(f"{independent_variable} and {dependent_variable} has a weak linear relationship")
            return 
```

```python
# Calculate Intercept
def intercept(slope: int, independent_variable: list, dependant_variable: list = None):
    
    if slope is not None and independent_variable is not None:
        
        if len(independent_variable) != dependant_variable:
            raise TypeError(f"Invalid Length: {len(independent_variable)}")
        
        elif not isinstance(slope, int):
            raise TypeError(f"Invalid slope: {type(slope)}")
        
        elif not isinstance(independent_variable, list):
            raise TypeError(f"Invalid array: {type(dependant_variable)}")
        
        for observation in independent_variable:
            if not isinstance(observation, int):
                raise TypeError (f"Invalid value: {observation}")
        
        summation_of_y = sum(dependant_variable)
        mean_of_y = summation_of_y / len(dependant_variable)

        summation_of_x = sum(independent_variable)
        mean_of_x = summation_of_x / len(independent_variable)

        intercept_value = mean_of_y - slope * mean_of_x
        print(f"Intercept value: {intercept_value}")

        return intercept_value
            
```

The function above takes in 'x' and returns a corresponding value 'y'
The slope and intercept are calculated using:
    slope, intercept, r, p, std_err = stats.linregress(x, y)


mymodel = list(map(myfunc, x))
This line of code is used to create a model, by applying myfunc to each value of x, hence producing a new list of coresponding 'y' values using the type converter list().




1. **Defining the Linear Model Function (`myfunc`):**
   - The function `myfunc(x)` represents the equation of a straight line: \( y = mx + c \), where:
     - \( x \) is the independent variable,
     - \( y \) is the dependent variable (predicted),
     - \( m \) is the slope (determined by the linear regression analysis),
     - \( c \) is the intercept (also determined by the linear regression analysis).
   - This function takes an \( x \) value and returns the predicted \( y \) value based on the linear model.

2. **Generating Predicted \( y \) Values (`mymodel`):**
   - Using the `map()` function, you apply the `myfunc` function to each value of the independent variable \( x \).
   - This generates a list of predicted \( y \) values based on the linear model for each corresponding \( x \) value.

3. **Creating the Scatter Plot:**
   - You then create a scatter plot using `plt.scatter(x, y)`, where `x` and `y` are the original data points.
   - This scatter plot shows the actual data points.

4. **Drawing the Regression Line:**
   - Finally, you draw the regression line using `plt.plot(x, mymodel)`.
   - This line represents the linear model's prediction for each \( x \) value.
   - It helps visualize how well the linear model fits the data.

By combining the scatter plot of the original data points with the regression line representing the linear model's predictions, you can visually assess the goodness of fit of the model to the data.
